```python
dataset = [11,10,12,14,12,15,14,13,15,17,19,107,10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
outliers=[]
def detect_outliers(data):

    threshold=3 #it means that is the it is 3rd standard deviation
    mean = np.mean(data)
    std =np.std(data)

    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```

##IQR Inter Quartile Range

*   Sort the data
*   calculate Q1 and Q3
*   Lower fence (Q1 - 1.5(iqr))
*   upper fence (Q3 + 1.5(iqr))





```python
dataset = sorted(dataset)
dataset
```

```python
q1,q3= np.percentile(dataset,[25,75])
```

```python
print(q1,q3)
```

```python
iqr = q3-q1
iqr
```

```python
lower_fence = q1 - (1.5*iqr)
upper_fence = q3 + (1.5*iqr)
```

```python
print(lower_fence,upper_fence)
```
